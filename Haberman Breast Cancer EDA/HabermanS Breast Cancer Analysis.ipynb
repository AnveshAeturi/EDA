<u><h1>1.Haberman's Survival Exploratory Data Analysis:</h1></u>  <br>
<h2>1.1 Introduction:</h2> <br> The dataset contains cases from a study that was conducted between 1958 and 1970 at the     university of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer. <br>

There are nearly 306 records or instances have beed entered in this dataset. They have considered nearly four features, which plays major role in the find the Survival Status of Patient after the Operation. The first Feature is <b>Age</b> of Patient at the time of operation,and second one is <b>Year</b> Patinent's year of operation which is 1900 - year for every particular Patient. And the third feature is <b>Axillary Nodes</b> which plays major role in dealing with cancer.  <br>

By combining all these features we can easily find weather the patient survived or not after the Operation happened. In this Haberman Dataset the Survial Status of Patient is given as two classes. <br>
<b>Class: 1</b> represents those who survived more than equal to 5 years after the operation happened. <br>
<b>Class: 2</b> represents those who survived less than 5 years after the operation happened.

<h1>1.6 CDF & PDF: </h1> <br>


```python
surv_1 = haberman.loc[haberman["Surv_Status"] == 1]
surv_2 = haberman.loc[haberman["Surv_Status"] == 2]
#print(haberman["Age "])
#print(haberman["Axil_nodes"])
#print(haberman["Surv_Status"])
```

<b>Surv_1:</b> Are those who survived more than five years after the operation has happened.<br>
<b>Surv_2:</b> Are those who survived less than five years . <br>
The Age varies form 30 to 77 for <b>surv_1</b> and 34 to 83 for <b>surv_2</b> categories. <br>


```python
counts, bin_edges = np.histogram(haberman["Age"], bins = 10, density = True)
pdf = counts/(sum(counts))
cdf  = np.cumsum(pdf)

#print("PDF: {}".format(pdf))
#print("Bin Edges: {}".format(pdf))

plt.plot(bin_edges[1:],pdf, label = "PDF of Age")
plt.plot(bin_edges[1:],cdf, label = "CDF of Age")

print("="*80)
counts, bin_edges = np.histogram(haberman["Age"], bins = 20, density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf, label = "PDF of Age")
#plt.grid()
plt.legend()
print("PDF: {}".format(pdf))
print("Bin Edges: {}".format(pdf))
plt.title("PDF & CDF for Age")
plt.xlabel("Age")
plt.ylabel("Probability")
plt.show()
```

<u><h3>Observations:</h3></u> <br>
<b> (10) </b> Around 15% of Patients have less than equal to Age 40, and 80% of Patients have age greater than 63.

```python
counts, bin_edges = np.histogram(haberman["Axil_nodes"], bins = 10, density = True)
pdf = counts/(sum(counts))
cdf  = np.cumsum(pdf)

print("PDF: {}".format(pdf))
print("Bin Edges: {}".format(pdf))

plt.plot(bin_edges[1:],pdf, label = "PDF of Axillary Nodes")
plt.plot(bin_edges[1:],cdf, label = "CDF of Axillary Nodes")
print("="*80)
counts, bin_edges = np.histogram(haberman["Axil_nodes"], bins = 20, density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf, label = "PDF of Axillary Nodes")

print("PDF: {}".format(pdf))
print("Bin Edges: {}".format(pdf))

plt.legend()
plt.title("PDF & CDF for Axillary Nodes")
plt.xlabel("Axillary Nodes")
plt.ylabel("Probability")
plt.show()
```
